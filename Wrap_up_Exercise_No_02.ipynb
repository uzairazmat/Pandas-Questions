```python
import pandas as pd   #import pandas as pd
```

```python
dict1 = {

        "name":["ahmad" , "mohsin" ,"husian " , "amna", "sana ", "ayesha"  ]  ,
        "marks" : [400,600, 300 , 900 , 788, 666],
        "gender": ["Male","Male","Male","Female", "Female", "Female"  ]

        }  #create dictionary



```

## 11. calculate average marks

```python
data = pd.DataFrame(dict1)  #convert dictionary into data frame
```

```python
data["marks"].mean()  #calculate mean value
```

## 12. Apply your own Funtion

```python
def half(num):
  """
  get a number and rerun there half
  4->2
  """
  return num/2
```

```python
data["marks"].apply(half)  #apply function passes one by one from the marks col to half funtion
```

## 13. Add new col in data

```python
data["half_marks"] = data["marks"].apply(half)   # calculating half and create new col
```

```python
data
```

## 14. Add a col of "male_femal" male  =1 , female = 0

```python
data["male_female"] = data["gender"].map({"male":1,"female":0})   #don't use this b/c male and female are unordered , categorical datahere just we use just for example
```

```python
data
```

## 15. drop half_marks	male_female col

```python
data.drop(["half_marks","male_female"],axis = 1 , inplace=True)  #use inplace methood to update
```
